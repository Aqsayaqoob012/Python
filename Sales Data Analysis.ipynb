### Sales Data Analysis

```python
'''  Tasks to Perform:
1. Create a new column Total_Sales by multiplying Units_Sold with Unit_Price.

2. Group the data by Region, Product, and Date. Calculate the following aggregated values:
Total Sales (sum)
Average Units Sold (mean)
Maximum Sales
Minimum Sales

3. Identify top performing products by sorting the products based on their total sales in ascending order. 
4. Identify top performing regions based on total sales, sorted in ascending order.
5.Handle missing values (if any) in the dataset.
(Optional) Normalize the Total_Sales column using z-score normalization with NumPy:
 ùëß=(ùë•‚àíùúá)ùúé z= œÉ(x‚àíŒº)

 
```

```python
import numpy as np
import pandas as pd
data = {
    'Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'C', 'A'],
    'Units_Sold': [10, 15, 10, 20, 25, 30, 15, 10, 20, 25],
    'Unit_Price': [100, 200, 150, 300, 200, 150, 300, 200, 300, 150]
}

# Calculate new columns (e.g., Total Sales = Units Sold √ó Unit Price).

df = pd.DataFrame(data)
df['Total_Sales'] = df['Units_Sold'] * df['Unit_Price']
print(f'Total Sales : {df['Total_Sales']} \n  \n')

''' Group data by Region, Product, or Date to calculate:
Total sales
Average units sold
Max/min sales per product or region'''

grouped = df.groupby(['Region' , 'Product' , 'Date']).agg(Total_Sales = ('Total_Sales' ,'sum' ) ,
                                                         Units_Sold=( 'Units_Sold' , 'mean'),
                                                          Max_Sales = ('Total_Sales' , 'max'),
                                                           Min_sales =('Total_Sales' , 'min'))
print(grouped)

#Filter and sort data to find top products or regions.

top_product = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending= True)
print(f' \n Top Product {top_product} \n')

top_region = df.groupby('Region')['Total_Sales'].sum().sort_values(ascending= True)
print(f' \n Top Region {top_region} \n')

#Normalize or transform sales data using NumPy (e.g., z-score normalization).

#Handle missing values if any.
df = df.isnull()
print(f' \n Missing values : {df} \n')

# normalize 
mean = df['Total_Sales'].mean()
std = df['Total_Sales'].std()
df['Z_normalize'] =( df['Total_Sales'] - mean) / std
print(f' \n Normalize : {df[['Total_Sales' ,'Z_normalize' ]]}')
```
