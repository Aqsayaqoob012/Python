```python
'''  Tasks:
✅ Step 1: Create data frame 
Display the first few rows

✅ Step 2: Calculate total and average marks for each student
Use NumPy to calculate total and average marks across all subjects

Add new columns Total and Average to the DataFrame

✅ Step 3: Identify the top 5 students
Sort the students based on their total marks (highest to lowest)
Display the names and total marks of the top 5 performers

✅ Step 4: Perform pass/fail analysis
Define a passing mark as 40 in each subject

Mark students as "Pass" if they passed all subjects, otherwise "Fail"

Add a new column Status to show the result

✅ Step 5: Calculate subject-wise average
Calculate and display the average marks for each subject
```

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Ali', 'Sara', 'Ahmed', 'Hina', 'Zain', 'Fatima', 'Usman', 'Ayesha', 'Bilal', 'Nida'],
    'Maths': [78, 90, 45, 66, 35, 88, 72, 95, 50, 62],
    'Physics': [65, 88, 50, 70, 30, 85, 64, 92, 55, 68],
    'Chemistry': [72, 85, 42, 60, 25, 80, 59, 88, 49, 61],
    'Biology': [70, 89, 48, 62, 33, 75, 60, 90, 55, 66],
    'English': [80, 95, 48, 72, 40, 90, 67, 93, 52, 74],
    'Urdu': [82, 91, 52, 68, 38, 86, 70, 94, 57, 79]
}
# Create the DataFrame
df = pd.DataFrame(data)
print(df.shape)
print(f' \n Display the first few rows :\n {df.head(3)} \n ')
#Step 2:

print('Calculate total and average marks for each student')
marks = df.iloc[: , 1:]
df['Total_Marks'] = np.sum(marks , axis = 1)
print(f' \n Total Marks :\n {df['Total_Marks']} ')
df['Average_Marks'] = np.mean(marks , axis = 1)   
print(f' \n Average Marks :\n {df['Average_Marks']} ')

#Step 3: 

top_5 = df.sort_values(by='Total_Marks' , ascending = False).head(5)
print('\n')
print('Identify the top 5 students :')
print(top_5[['Name' , 'Total_Marks' ]] )

 #Step 4 :

Pass_Marks = 40 
df['Status'] = np.where(
    (df['Maths'] > Pass_Marks) &
    (df['Physics'] > Pass_Marks) &
    (df['Chemistry'] > Pass_Marks) &
    (df['Biology'] > Pass_Marks) &
    (df['English'] > Pass_Marks) &
    (df['Urdu'] > Pass_Marks),
    'Pass' , 'Fail'
)
print('\n')
print('pass/fail analysis :')
print(df[['Name' , 'Status']])
 #Step 5 :
print('\n')
print('Calculate subject-wise average')
print(df.agg ({'Maths' : 'mean' , 'Physics' : 'mean' , 
               'Chemistry' : 'mean' , 'Biology' : 'mean' ,
               'English' : 'mean' , 'Urdu' : 'mean'     }))

```
